```python
# code to display word length greater than given number
def bbbb(b):
    
    j=input().split()
    c=[len(j[i]) for i in range(len(j))]
    for i in range(len(c)):
        if int(c[i])>b:
            print(j[i])
            
    return
```

```python
bbbb(5)
```
